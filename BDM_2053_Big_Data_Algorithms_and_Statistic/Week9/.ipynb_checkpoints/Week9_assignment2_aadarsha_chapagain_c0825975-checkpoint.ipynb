```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('./zoodata.csv')
```

```python
df.head()
```

```python
df.columns =['name', 'hair', 'feathers', 'eggs', 'milk', 'airbone', 'aquatic', 'predator', 'toothed', 'backone', 'breathes', 'venomous', 'fins', 'legs', 'tail', 'domestic', 'catsize', 'type'
]
```

```python
df.head(200)
```

```python
print("Shape",df.shape)
```

```python
df.describe()
```

```python
x =df.iloc[:, :-1]
y =df.iloc[:, -1]
```

## Encode our target variable into a nummerical vairable encode label encode command

```python

from sklearn.preprocessing import OneHotEncoder

encoded_x = OneHotEncoder().fit_transform(x).toarray()
print(encoded_x)
```

## Split into Traning & Testing Sets 

```python
 from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(encoded_x, y, test_size=0.30, random_state=123)
```

```python
x_train
```

## Create Classifier object

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,   hidden_layer_sizes=(5, 2), random_state=1)
```

## Train MODEL

```python
clf.fit(x_train, y_train)
```

## Make Predictions

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

## CONFUSION Matrix

```python
confusion_matrix(y_test, y_pred)
```

### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
