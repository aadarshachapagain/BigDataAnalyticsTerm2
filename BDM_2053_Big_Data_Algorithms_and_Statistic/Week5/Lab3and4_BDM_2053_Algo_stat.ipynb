```python
# Probablity Distributions
#import libraries
import numpy as np
import pandas as pd
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
```

```python
# Uniform Distribution - each value within a certain range is equally likely to occur
uniform_data= stats.uniform.rvs(size=100000, # Generate 100000 numbers
                                   loc = 0,     # From 0
                                   scale =10)   # To 10

pd.DataFrame(uniform_data).plot(kind="density",  # Plot the distribution
                               figsize=(9,9),
                               xlim=(-1,11),
                               title="Uniform Distribution")
```

```python
#  Normal Distribution
'''The Normal Distribution/Gaussian Distribution - Symmetric Bell Shaped Curve
A normal Distribution is defined by its center -mean and spread-standard devation
the bulk of the observations generated from a normal distribution lie near the
mean, which lies at the exact center of the distribution,68% of the data lies 
within 1 standard deviation  of the mean, 95% lies within 2 standard deviations 
and 99.7% lies within 3 standard deviations. many common statistical tests assume
distributions are normal.'''
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

domain = np.linspace(-2,2,1000)
plt.plot(domain, norm.pdf(domain,0,1))
plt.title("Standard Normal")
plt.xlabel("Value")
plt.ylabel("Dentsity")
plt.show()
```

```python
# Binomial Distribution
'''The Binomial Distribution - The binomial distribution tells how likely it is
achieve a given number of sucesses in n trials of the experment'''

fair_coin_flips = stats.binom.rvs(n=10,   # number of flips
                                 p=0.5,   # success porbability
                                 size=10000) # Number of trails 

print(pd.crosstab(index ="counts", columns = fair_coin_flips))

pd.DataFrame( fair_coin_flips). hist(range=(-0.5, 10.5), bins =11);
```

```python
# Poission Distribution
'''The Poisson Distribution 
certain number of sucesses with in a time interval , eg: number of emails/text msg
you'd expect to receive in a week.'''
random.seed(12)
arrival_rate_1 = stats.poisson.rvs(size=10000, # Generate Poisson data
                                    mu=1) # Average arroval time 1
#Print table of Counts
print(pd.crosstab(index="counts", columns= arrival_rate_1))

# Plot histogram
pd.DataFrame(arrival_rate_1).hist(range=(-0.5, max(arrival_rate_1) +0.5)
                                             , bins =max(arrival_rate_1) +1)
```

```python
#Confidence Interval
'''Confidence Intervals
point Estimates: The sample mean is usually not exactly same as the population
mean.'''

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import math

np.random.seed(10)
population_ages1 = stats.poisson.rvs(loc=18, mu=35, size=150000)
population_ages2 =  stats.poisson.rvs(loc=18, mu=10, size= 100000)
population_ages = np.concatenate((population_ages1, population_ages2))

population_ages.mean()                  

np.random.seed(6)
sample_ages = np.random.choice(a = population_ages, size = 500) # sample 1000 values

print (sample_ages.mean()) # show sample mean

population_ages.mean() - sample_ages.mean() # check difference between means


''' Confidence Intervals:
A point estimate can give you a rough idea of a population parameter like the
mean, but estimates are prone to error and taking multiple samples to get improved
estimates may not be feasible. a confidence interval is a range of value above
and below a point estimate that captures the true population parameter at some
predetermined confidence level. 

margin of error , ctritical value '''

np.random.seed(10)
sample_size = 1000 
sample = np.random.choice(a = population_ages, size = sample_size)
sample_mean = sample.mean()

z_critical = stats.norm.ppf(q= 0.975) # Get the z-critical value

print("z-critical value:")  # get the z-critical value

print(z_critical)

pop_stdev = population_ages.std() # get the population standard deviation
 
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean  + margin_of_error)

print("Confidence interval:")
print(confidence_interval)


# t-test

np.random.seed(10)
sample_size = 1000 
sample = np.random.choice(a = population_ages, size = sample_size)
sample_mean = sample.mean()

z_critical = stats.norm.ppf(q= 0.975) # Get the z-critical value

print("z-critical value:")  # get the z-critical value

print(z_critical)

pop_stdev = population_ages.std() # get the population standard deviation
 
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean  + margin_of_error)

print("Confidence interval:")
print(confidence_interval)


# stats.t.interval(alpha = 0.95, #Confidence Level
#                     df=24, # Degrees of Freedom
#                     loc=sample_mean, # Sample mean
#                     scale = sigma) # Standard Deviation estimate
```
