```python
#drop unimportant variables
clean_cars = cars.drop(['MSRP', 'Number of Doors'], axis=1)
clean_cars
```

```python
#drop duplicate rows- if any
cars = clean_cars.drop_duplicates(keep='first')
cars
```

```python
#shape of the data
cars.shape
```

```python
#summary statistics
cars.describe()
```

```python
#sorting a variable
cars_sort= cars.sort_values(by='city mpg', ascending=False)
cars_sort.head()

```

```python
#replace space in column name
cars.columns = cars.columns.str.replace(' ', '')
cars
```

```python
#histogram of continious numerical variable
import matplotlib.pyplot as plt
import seaborn as sns
num_bins=10
plt.hist(cars['citympg'],num_bins)
```

```python
#probability density function-PDF
sns.distplot(cars['citympg'],bins=10)
```

```python
#count by category - cross tabulate
make_dist=cars.groupby('Make').size()
make_dist
```

```python
#Distribution of Categorical variable
make_dist.plot(title='Make')
```

```python
#select all numerical varibales
cars_num = cars.select_dtypes(include=['float64', 'int64'])
# all varibales ploting
cars_num.hist(bins=20)
```

```python
#correlation with the varibale of interest
cars_corr = cars_num.corr()['citympg'][:-1]
cars_corr
```

```python
#correlation plots using 'pairplot'
for i in range(0, len(cars_num.columns),5):
    sns.pairplot(cars_num, y_vars=['citympg'],x_vars=cars_num.columns[i:i+5])
```
