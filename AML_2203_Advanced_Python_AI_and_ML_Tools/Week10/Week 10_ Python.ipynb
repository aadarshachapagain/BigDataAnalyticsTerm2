```python
import numpy as np
import skimage
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = io.imread("image.png")
```

```python
print(type(img))
print(img.shape)
```

```python
plt.imshow(img)
```

```python
from skimage.transform import resize
# resize image  
resized = resize(img, (400, 400))
plt.imshow(resized)
print(resized.shape)
```

```python
img_crop = img[80:280, 50:288]
plt.imshow(img_crop)
print(img_crop.shape)
```

Savings image

```python
io.imsave('image_2.png',img)
io.imsave('image_3.png',resized)
io.imsave('image_4.png',img_crop)
```
