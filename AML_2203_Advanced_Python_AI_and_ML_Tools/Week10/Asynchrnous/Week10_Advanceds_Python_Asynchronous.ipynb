```python
#Importing required Libraries

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#read Image using imread function
img = cv2.imread("cat_image.png")
```

```python
#printing type and shape of the array
print(type(img))
print(img.shape)
```

```python
#Plottig the result
plt.imshow(img)
```

```python
#Convert Image from BRG to RGB using cvtColor function
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#Image Resize
scale = 20
new_width = int(img_rgb.shape[1] * scale / 100)
new_height = int(img_rgb.shape[0] * scale / 100)
dim = (new_width,new_height)


resized = cv2.resize(img_rgb, dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized)
print(resized.shape)
```

```python
#Cropping the Image
img_crop = img_rgb[550:990, 100:800]
plt.imshow(img_crop)
print(img_crop.shape)
```

```python
#Saving Image
cv2.imwrite('pro_1.png',img)
cv2.imwrite('pro_2.png',resized)
cv2.imwrite('pro_3.png',img_crop)
```

##3.Using Pillow

```python
#Improting Libraries
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#Read the Image
img_2 = Image.open("cat_image.png")
```

```python
#Printing the type and the shape of the image
print(type(img_2))
print(img_2.size)
```

```python
#plotting the result
plt.imshow(img_2)
```

```python
#Resizing the image
size = (100,100)

resized  = img_2.resize(size)
plt.imshow(resized)
print(resized.size)
```

```python
#Cropping the Image
img_crop = img_2.crop((170,500, 600,900))
plt.imshow(img_crop)
print(img_crop.size)
```

```python
#Saving Image
img_2.save('pr0_4.png')
resized.save('pr0_5.png')
img_crop.save('pr0_6.png')
```
